INSIGHTS
-There are missing values in 3 columns ,Age,Cabin,Embarked
-As there are many missing values in Cabin column ,its better to drop the column
-The missing values in Age and Embarked column has to be filled
-'Name'and 'Ticket'is not contributing to target 'Survived',so we remove it from the dataset


```python
#Dropping the columns:'Cabin','Name','Ticket'
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.drop('Name',axis=1,inplace=True)
df.drop('Ticket',axis=1,inplace=True)
```

```python
df
```

```python
#Filling of null values
df.value_counts('Embarked')
```

Most of them have embarked from S ,we can fill the missing values with 'S'

```python
df['Embarked']=df['Embarked'].fillna('S')
```

```python
#Checking for outliers
plt.boxplot(df['Fare'])
plt.title('Box plot of fare')
```

```python
import numpy as np
Q1=np.percentile(df['Fare'],25)
Q2=np.percentile(df['Fare'],50)
Q3=np.percentile(df['Fare'],75)
print (Q1)
print(Q2)
print(Q3)
```

```python
IQR=Q3-Q1
low_lim=Q1-1.5*IQR
Up_lim=Q3+1.5*IQR
print(low_lim)
print(Up_lim)
```

```python
outlier=[]
for x in df['Fare']:
    if((x>Up_lim)or(x<low_lim)):
        outlier.append(x)
outlier    
```

```python
#Preparing a model
Y=df['Survived']
X=df.drop('Survived',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=42,test_size=0.2)

```

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
model=lr.fit(X_train,Y_train)
predictions=model.predict(X_test)



```

```python
from sklearn.metrics import mean_squared_error
print('MSE is:',mean_squared_error(Y_test,predictions))
```

```python
from sklearn.metrics import r2_score
print('R squared value is :',r2_score(Y_test,predictions))
```
