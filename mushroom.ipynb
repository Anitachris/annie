```python
#Dropping the 'veil-type'column
df=df.drop(["veil-type"],axis=1)
df
```

# Label Encoding


```python
#Changing the datatype in to category
df=df.astype('category')
df.dtypes
```

```python
#Using label encoder
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
for column in df.columns:
    df[column]=lb.fit_transform(df[column])
df.head()
```

```python
train_x = df.drop(["class"], axis=1).values
train_y=df['class']
test_x= df.drop(["class"], axis=1).values
test_y =df['class']
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(train_x,train_y,test_size=0.3,random_state=2)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

## lr = LogisticRegression(solver="lbfgs")
lr = LogisticRegression(solver="liblinear")
lr.fit(train_x,train_y)

print("Test Accuracy: {}%".format(round(lr.score(test_x,test_y)*100,2)))

```

```python
y_pred = lr.predict(test_x)

y_pred
```

```python
from sklearn.metrics import confusion_matrix

#Generate the confusion matrix
cf_matrix = confusion_matrix(test_y, y_pred)

print(cf_matrix)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
model=KNeighborsClassifier(n_neighbors=3)
model.fit(train_x,train_y)
prediction=model.predict(test_x)
print('The accuracy of the KNN  is',metrics.accuracy_score(prediction,test_y))
```

```python
y_pred = model.predict(test_x)

y_pred
```
